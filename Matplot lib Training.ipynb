```python
import pandas as pd    #pandas allows you to turn your project into a data frame 
import numpy as np  #allows for flexible interaction of your data 
%matplotlib inline
import matplotlib.pyplot as plt   #ths is a visual representation of our data
import seaborn as sns
```

Matplotlib

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'g*')
plt.xlabel('Output')
plt.ylabel('Price')
plt.title('Supply')
```

```python
#plt.subplot(nrows, ncol, plot_number)
plt.subplot(1,2,1)
plt.figure(figsize=(3, 3))
plt.plot(x, y, 'r')
plt.subplot(1,2,2)
plt.plot(y, x, 'g')
```

#OBJECTORIENTED

```python
#creating a figure
fig = plt.figure() 
#Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height
#plot the figure on the axes
axes.plot(x, y, 'b')
axes.set_xlabel('Quantity')
axes.set_ylabel('Price')
fig.savefig('plot.png')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = data)
```

```python
sns.countplot(x='sex', data= data, palette = 'viridis')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = data, palette = 'rainbow')
```
